```python
import pandas as pd

def validate_column(df, column_name, output_csv="validated_data1.csv"):

    # Convert column to uppercase
    df[column_name] = df[column_name].astype(str).str.upper()

    # Validate length (10 characters)
    df['Flag_For_10_Digit_Count'] = df[column_name].apply(lambda x: 'Valid' if len(x) == 10 else 'Invalid')

    # Save to CSV
    df.to_csv(output_csv, index=False)
    print(f"Validation complete. Updated file saved as '{output_csv}'.")

    return df

# Example usage
file_path = "/content/Form10BD_Donations_Child Help Foundation 31.01.2025.xlsx"  # Replace with actual file path
df = pd.read_excel(file_path)

# Call the function for Column D and save to CSV
df = validate_column(df, 'Unique Identification Number', 'validated_data1.csv')

# Display the updated DataFrame
print(df)
```
